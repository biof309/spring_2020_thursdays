## Final informational check

The dollar sign in bash denotes a variable. $0 is a special variable that contains the name of the current shell being used:

```python
%%bash
echo $0
```

We can achieve the above command by using a system call in ipython. Using $0 directly gives a confusing result but we can assign this to a string in python and use braces to expand it to create the command we want to run:

```python
variable_representing_the_current_shell = '$0'
!echo {variable_representing_the_current_shell}
```

Finally, it is useful to look at the PATH variable in bash. This will tell us where we find software. If we are looking for software and it is not on our path (stored in teh PATH variable) we are in trouble...

```python
%%bash 
echo $PATH
```
